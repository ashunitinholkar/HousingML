```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Loading data
dataset = pd.read_csv('housing.csv')
```

```python
dataset.hist(bins = 50, figsize=(20,19))
plt.show()
```

```python
# Plotting data

X = dataset.iloc[:,:8].values
y = dataset.iloc[:,8].values

plt.plot(X,y)
plt.title('Housing Features Vs Housing Price')
plt.xlabel('Features')
plt.ylabel('Housing Price')
plt.show()
```

```python
#Converting 1D to 2D
t_size = int(len(dataset))
X,y = np.arange(t_size).reshape(int(t_size/2),2),range(int(t_size/2))

#Splitting Data into Training and Testing
trainX, testX, trainY, testY = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
#Plotting training set
plt.plot(trainX,trainY)
plt.title('Training')
plt.xlabel('trainX')
plt.ylabel('trainY')
plt.show()

#Performing Linear Regression
linearReg = LinearRegression()
linearReg.fit(trainX,trainY)
```

```python
#Testing model on testset
ypred = linearReg.predict(testX)

#Plotting testing set
plt.plot(testX,testY,color='blue')
plt.title('Testing')
plt.xlabel('testX')
plt.ylabel('testY')
plt.show()

#Plotting trained testing set
plt.plot(testX,ypred,color='red')
plt.title('Trained Test')
plt.xlabel('testX')
plt.ylabel('ypred')
plt.show()
```

```python
#calculating accuracy
p = list()
sz = int(len(testY))
for i in range(sz):
    if ypred[i] == testY[i]:
        p.append(1)
    else:
        p.append(0)
accuracy = sum(p)*100/sz

print("Accuracy of Testing Set Is: ",accuracy,"%")
```
